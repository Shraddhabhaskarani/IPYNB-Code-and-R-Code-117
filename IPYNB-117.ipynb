```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd # import pandas

# Sample data (replace this with your actual data)
# Make sure both lists have the same number of elements
data = {'Customers': [555,625,821,1498,559,589,1414,833,687,681,1236,962,568,710,766,979,946,936,718,974,582,633,560,1082,1586,611,1263,663,737,625,645,635,915,1162,777,1251,888,608,929,678],
        'Sales': [523,6064,8314,13995,4822,5651,15344,8492,8565,7185,10457,8959,8821,6544,9191,10231,8430,10071,8234,959,9593,9515,6566,7273,14190,14180,7818,13213,7301,9411,4776,7248,6396,10789,11144,12412,12422,9091,7434,8336]}
df = pd.DataFrame(data) # Create a DataFrame named df

# Assuming 'Sales' and 'Customers' are columns in your DataFrame 'df'

X = df[['Customers']]  # Features (independent variable)
y = df['Sales']  # Target (dependent variable)

# Split the data into training and testing sets
# Reduced test_size to allow for splitting with smaller datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Create a linear regression model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Create a scatter plot of Sales vs Customers
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Customers', y='Sales', data=df)

# Plot the regression line
plt.plot(X_test, y_pred, color='red', label='Linear Regression')

plt.xlabel('Customers')
plt.ylabel('Sales')
plt.title('Linear Regression: Sales vs Customers')
plt.legend()
plt.show()
```
